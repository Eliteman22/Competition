```python
import pandas as pd
from battlefin import BattleFin
```

```python
# Initialize the BattleFin Wrapper
bf = BattleFin()
```

```python
# Pull the fundamental data into the environment
fundamentals = bf.getFullDataset('fundamentals_competition',returns='pandas')
```

```python
fundamentals.head()
```

```python
# Pull a certain ticker from the fundamental data
NFLX_fundamentals = bf.get('fundamentals_competition',search_column='ticker',search_value='NFLX')
```
